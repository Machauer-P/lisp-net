# 2D Data Generation Pipeline

This notebook provides an interactive interface for generating 2D datasets with integrated offsets for comparison benchmarks (e.g., UniverSeg).

```python
import os
import sys
from pathlib import Path

# Setup project root and paths
notebook_dir = Path().resolve()
project_root = notebook_dir.parent.parent
if str(project_root) not in sys.path:
    sys.path.insert(0, str(project_root))

from data.test_data.generate_2d_test_data import gen_save_ds
from data.DataLoader_npz import DataLoader_npz
from data.DataGenerator import DataGenerator
from data.test_data.ds_handler_2d import load_tf_dataset_2D
```

## 1. Parameters Configuration

```python
settings = {
    "input_files": [
        "data/test_data/HanSeg_MRI.npz",
        "data/test_data/HanSeg_CT.npz"
    ],
    "output_dir": "data/test_data/2d/offset_5",
    "ds_name": "combined",
    "num_ds": 10,        # Number of datasets to generate
    "offset": 5,         # Slice offset for prompt/target pairs
    "max_points": 100,   # Max query points per dataset
    "len_p_set": 16      # Number of support samples
}

# Resolve absolute paths
input_paths = [str(project_root / p) for p in settings["input_files"]]
output_path = str(project_root / settings["output_dir"])

print(f"Inputs: {len(input_paths)} files")
print(f"Output directory: {output_path}")
```

## 2. Initialize and Run Generation

```python
print("Initializing components...")
dl = DataLoader_npz(input_paths, val_size=0.0)
dg = DataGenerator(dl)

print("Starting generation flow...")
gen_save_ds(
    dg=dg,
    path=output_path,
    ds_name=settings["ds_name"],
    offset_val=settings["offset"],
    num_ds=settings["num_ds"],
    max_data_points=settings["max_points"],
    len_p_set=settings["len_p_set"]
)
```

## 3. Quick Verification
Load one sample to ensure offsets are correctly integrated.

```python
test_filename = f"0_{settings['ds_name']}"
ds = load_tf_dataset_2D(test_filename, path=output_path, include_offset=True)

for x, y, p, offset in ds.take(10):
    print(f"Successfully loaded query sample with integrated offset: {offset.numpy()}")
    print(f"X shape: {x.shape}, Y shape: {y.shape}, Support P shape: {p.shape}")
```
